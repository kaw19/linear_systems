<h2 style="color:blue; background-color:#F3F2F9">Fundamentação Teórica</h2>

#### Série Trigonométrica:

Um sinal periódico $x(t)$, com período fundamental $T_0 = 2\pi/\omega_0$, atende à seguinte expressão:  
$$x(t)=x(t+T_0) \tag{1}$$
  
Fourier mostrou, em 1823, que o sinal $x(t)$ pode ser representado por uma soma de infinitas senóides, dado por:
$$x(t)=\frac{a_0}{2}+\sum_{k=1}^\infty a_k.cos(k\omega_0 t)+b_k.sen(k\omega_0 t) \tag{2}$$  
onde:
$$a_0=\frac{2}{T_0}\int_{<T_0>} x(t)dt \tag{3}$$  
$$a_k=\frac{2}{T_0}\int_{<T_0>} x(t).cos(k\omega_0 t)dt \tag{4}$$  
$$b_k=\frac{2}{T_0}\int_{<T_0>} x(t).cos(k\omega_0 t)dt \tag{5}$$  

#### Série Exponencial Complexa:  

A Série de Fourier Exponencial Complexa é dada por:

$$x(t)=\sum_{-\infty}^\infty D_k.e^{-jk\omega_0 t} \tag{6}$$  
onde:  
$$D_k=\frac{1}{T_0}\int_{<T_0>} x(t).e^{-jk\omega_0 t}dt,\;\;\;\;\;-\infty<k<\infty \tag{7}$$ 

<h2 style="color:blue; background-color:#F3F2F9">Procedimento Prático</h2>

<ol><li>Dado o sinal dente-de-serra, periódico no tempo contínuo: $x(t)=\begin{cases} & t \text{, }-\pi<t<\pi \\ & x(t+2k\pi)=x(t) \text{, } k \in \mathbb{N} \end{cases}$.  
    
Os coeficientes da Série Trigonométrica de Fourier, conforme calculado (slides), são:     
$$a_0=a_k=0 \;\;\;\text{e}\;\;\; b_k=2\frac{(-1)^{k+1}}{k}$$

<ol><li type="a">Trace quatro períodos dos sinais, original e sintetizado pela Série Trigonométrica de Fourier, usando apenas os quatro primeiros harmônicos ($0 \le k \le 3$).</li>

```python
from numpy import arange, pi, sin, zeros
from matplotlib.pyplot import plot, grid, title, xlabel, axhline, axvline, legend, figure

T0 = 2*pi                               # período (s)
w0 = 2*pi/T0                            # frequência angular (rad/s)
t = arange(-2*T0,2*T0,0.01)             # tempo (s) --> abscissas
x = (t - pi) % (2.*pi) - pi             # amplitude do sinal x(t) --> ordenadas

figure(figsize=(12, 5)); plot(t,x,'b',linewidth=2) 
title('Sinal Dente-de-Serra'); xlabel('tempo'); ylabel('Amplitude'); grid('on')    
k = arange(3)+1
b = 2.*(-1)**(k+1)/k                    # coeficientes dos três primeiros Senos 
s = b[0]*sin(1.*w0*t) + b[1]*sin(2.*w0*t) + b[2]*sin(3.*w0*t)   # sinal sintetizado com 3 harmônicos

plot(t,s,color='r')                     # sinal sintetizado com 4 harmônicos
axhline(color='k'); axvline(color='k')  # eixos: horizontal e vertical
legend(('Original','Sintetizado'));
```

b. Idem ao item anterior, porém usando os 10 primeiros harmônicos da Série de Fourier.

```python
figure(figsize=(12, 5)); plot(t,x,'b',linewidth=2)
title('Sinal Dente-de-Serra'); xlabel('tempo'); ylabel('Amplitude'); grid('on')    
faixa = arange(10) + 1
b = 2.*(-1)**(faixa+1)/faixa                   # coeficientes dos três primeiros Senos 
s = 0.
for k in faixa:
    s += b[k-1]*sin(k*w0*t)                    # acumulação dos 10 harmônicos do sinal sintetizado

plot(t,s,color='r')                            # sinal sintetizado com 4 harmônicos
axhline(color='k'); axvline(color='k')         # eixos: horizontal e vertical
legend(('Original','Sintetizado'));
```

```python
%matplotlib inline
# Traçado da Resposta em Frequência de um circuito RC Série
from numpy import arange, abs, angle, pi
from matplotlib.pylab import plot, grid, title, subplot, xlabel, ylabel, figure

w = arange(0,5000*pi,10*pi)     # base de tempo
R = 1e3; C = 250e-9             # fc = 4000 rad/s = 636,2 Hz
H = 1. / (1. + 1j*w*R*C)        # Resposta em Frequência de um circuito RC série

figure(figsize=(12, 5)); subplot(1,2,1); plot(w,abs(H)); grid('on'); 
title(u'Resp. em Frequência'); xlabel('$\omega$ (rad/s)'); ylabel('Magnitude: $|H(\omega)|$')

subplot(1,2,2); plot(w,angle(H)); grid('on') 
title(u'Resp. em Frequência'); xlabel('$\omega$ (rad/s)'); ylabel('Fase: $\\angle{H(\omega)}\;\;$(rad)');
```
